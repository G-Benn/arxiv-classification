```python
import pandas as pd
import re
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
info_df = pd.read_pickle('test_src/info_df.pkl')
reduced_categories_df = pd.read_pickle('test_src/full_reduced_df.pkl')
expanded_categories_df = pd.read_pickle('test_src/full_expanded_df.pkl')
```

```python
categories = list(reduced_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).columns.values)
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))

ax= sns.barplot(categories, reduced_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).sum())

plt.title("Papers in each primary category", fontsize=24)
plt.ylabel('Number of papers', fontsize=18)
plt.xlabel('Paper Categories', fontsize=18)
plt.xticks(rotation=-50)

plt.show()
```

```python
categories = list(expanded_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).columns.values)
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))

ax= sns.barplot(categories, expanded_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).sum())

plt.title("Papers in each subcategory", fontsize=24)
plt.ylabel('Number of papers', fontsize=18)
plt.xlabel('Paper Categories', fontsize=18)
plt.xticks(rotation=-50)

plt.show()
```

```python
# Categories and subcategories per paper (histogram)
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))

ax= sns.distplot(expanded_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).sum(axis=1))

plt.title("Number of subcategories per paper", fontsize=24)
plt.ylabel('Number of papers', fontsize=18)
plt.xlabel('Number of subcategories', fontsize=18)

plt.show()
```

```python
# Categories and subcategories per paper (histogram)
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))

ax= sns.distplot(reduced_categories_df.drop(columns=['int_paper_id','text','length','paper_id','abstract','authors','title']).sum(axis=1))

plt.title("Number of categories per paper", fontsize=24)
plt.ylabel('Number of papers', fontsize=18)
plt.xlabel('Number of categories', fontsize=18)

plt.show()
```

```python
# Categories and subcategories per paper (histogram)
info_df['length'] = info_df['text'].apply(len)
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))

ax= sns.distplot(info_df['length'],bins=20)

plt.title("Number of words in paper", fontsize=24)
plt.ylabel('Number of papers', fontsize=18)
plt.xlabel('Number of words', fontsize=18)

plt.show()
```
