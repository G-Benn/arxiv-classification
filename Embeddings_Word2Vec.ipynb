Experimentation and initial test runs to generate the embeddings for the full text of articles

```python
import pandas as pd
import gensim as gs
import sklearn as sk
import numpy as np
from gensim.models import Word2Vec
import gensim
import re
import nltk
```

```python
import gensim
assert gensim.models.doc2vec.FAST_VERSION > -1
```

```python
info_df = pd.read_pickle('test_src/info_df.pkl')
reduced_categories_df = pd.read_pickle('test_src/full_reduced_df.pkl')
expanded_categories_df = pd.read_pickle('test_src/full_expanded_df.pkl')
```

```python
processed_article = info_df['text'][1].lower()  # put it all to lowercase
processed_article = re.sub('[^a-zA-Z]', ' ', processed_article )  # Remove non-alphabet characters
processed_article = re.sub(r'\s\w{1,2}\s', '', processed_article) # Remove 1|2-instance 'words'
processed_article = re.sub(r'\s+', ' ', processed_article) # Remove multiple spaces


# Preparing the dataset
all_sentences = nltk.sent_tokenize(processed_article)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

# Removing Stop Words
from nltk.corpus import stopwords  
for i in range(len(all_words)):  
    all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english')]
```

```python
print(all_words)
```

```python
word2vec = Word2Vec(all_words, min_count=2)
```

```python
vocabulary = word2vec.wv.vocab  
print(vocabulary) 
```

```python
v1 = word2vec.wv['planets']  
print(v1)
```

```python
sim_words = word2vec.wv.most_similar('heat')  
print(sim_words)
```
